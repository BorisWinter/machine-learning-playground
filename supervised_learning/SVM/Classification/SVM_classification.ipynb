## SVM classification
### Imports

```python
import pandas as pd
import numpy as np
import sklearn.datasets as data
from sklearn.model_selection import train_test_split, cross_validate
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix
from sklearn.inspection import DecisionBoundaryDisplay
import matplotlib.pyplot as plt
import seaborn as sns
```

### Load sample dataset

```python
# Multi-class data
# X, y = data.load_iris(return_X_y=True)
# X_train, X_test, y_train, y_test = \
#     train_test_split(X, y, test_size=0.20, random_state=42)

# Binary data
X, y = data.load_breast_cancer(return_X_y=True)

# Limit to two features
# X = X[:,:2]

X_train, X_test, y_train, y_test = \
    train_test_split(X, y, test_size=0.20, random_state=42)
```

### Visualize high-dimensional data using t-SNE 

```python
from sklearn.manifold import TSNE

X_embedded = TSNE(n_components=2, learning_rate='auto',
    init='random', perplexity=15).fit_transform(X)

plt.scatter(X_embedded[:, 0], X_embedded[:, 1], c=y, edgecolor="k")
```

### Instantiate algorithms

```python
model = SVC(gamma='auto')
clf = make_pipeline(StandardScaler(), model)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
```

### Plot classification performance confusion matrix

```python
# Generate confusion matrices
_matrices = []
# _matrices.append(confusion_matrix(y_test, y_pred_log_reg, normalize="true"))
_matrices.append(confusion_matrix(y_test, y_pred, normalize="true"))

class_names=[0,1] # name  of classes
fig, ax = plt.subplots(1,2, figsize=(10, 4), dpi=500)
plt.tight_layout()
plt.suptitle('Confusion matrix', y=1.1)
tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names)
plt.yticks(tick_marks, class_names)

for i, m in enumerate(_matrices):
    # create heatmap
    sns.heatmap(pd.DataFrame(m), ax=ax[i],
                annot=True, cmap="YlGnBu", fmt='g')

    # ax.xaxis.set_label_position("top")
    ax[i].set_ylabel('Actual label')
    ax[i].set_xlabel('Predicted label')
```
